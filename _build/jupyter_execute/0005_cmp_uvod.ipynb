Koncept stavebnice je navrhnutý tak, aby bola dostupná každého mladého konštruktéra. Všetky podklady a konštrukčné návrhy sú open-source, pre tlač dielov je možné využiť bežne dostupné 3D tlačiarne, pre tlač dielov je použitý recyklovateľný materiál PLA, ako spojovacie prvky sú použité skrutky a diely bežne dostupné v každom železiarstve, pre riadenie elektronických častí je možné využiť dostupné moduly a diely z platformy Arduino. 

## <font color='purple'> História </font>

`STEMFIE-X` je odvodená z konštrukčnej stavebnice pre deti  [*STEMFIE*](https://stemfie.org/). Myšlienka stavebnice je založená na návrhu a následnej 3D tlači univerzálnych komponentov a spojovacích prvkov a konštrukcii vlastných modelov. Modely je možné zostavovať z náverhov hotových sád dielov, ktoré sa vytlačia na 3D tlačiarni, na návrh a konštrukciu vlastných modelov je možné použiť grafické programy a konštrukčné CAD systémy ako FreeCAD, Blender a Tinkercad.

Stavebnica *STEMFIE-X* využíva základné princípy projektu *STEMFIE* s úpravami pre použitie ako platforma pre vytváranie zložitejších štruktúr a modelov s použitím mechanických a elektronických prvkov - motorov, servopohonov, elektronických obvodov a riadenia mikrokontrolérmi. Stavebnica je primárne určená pre polytechnickú výučbu na základných a stredných školách a prvých semestrov vysokých škôl pri výučbe predmetov z oblasti programovania, parametrického modelovania, CAD systémov, konštrukcií strojov a mechanizmov. 

Základným rozdielom medzi *STEMFIE-X* a *STEMFIE* je veľkosť jednotka (BU), ktorá určuje vzdialenosti medzi montážnymi otvormi, pôvodná veľkosť 12.5mm v *STEMFIE* je upravená na veľkosť 10mm a montážne otvory majú priemer 4 mm. Tieto modifikácie umožňujú použiť ako spojovací materiál štandardné skrutky M4 a diely stavebnice sú rozmerovo kompatibilné s kovovými dielmi stavebníc *Meccano*, *Merkur*, *eitech* a podobnými.

Základnými prvkami stavebnice sú spojky a bloky, ktoré je možné modifikovať v násobkoch BU, rotačné diely tvoria kladky a kolesá, pre pokročílé konštruekcie je možné využiť ozubené kolesá a pomocné prvky pre montáž motorov, serv a senzorov.
